```python
# VIDEO (pop-up, abort with "q") - # Visualize the results - https://docs.ultralytics.com/modes/predict/#plotting-results

# load the trained model
model = YOLO(r"C:\Users\marku\Desktop\MLDE22\MLDE_Kevat_2024\Project_ROBOTICS_AI\robotics_project\python\object_detection_model\best.pt")

# Open the video file
video_path = r"C:\Users\marku\Desktop\MLDE22\MLDE_Kevat_2024\Project_ROBOTICS_AI\robotics_project\python\object_detection_model\test_vids\PXL_20240322_073139460.TS.mp4"
cap = cv2.VideoCapture(video_path)

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLOv8 inference on the frame
        results = model(frame)

        # Visualize the results on the frame
        annotated_frame = results[0].plot()

        # Display the annotated frame
        cv2.imshow("YOLOv8 Inference", annotated_frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```
